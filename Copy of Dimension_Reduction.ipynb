```python
df=pd.read_csv('winequality_white.csv')
```

```python
y=df.iloc[:,-1] #independent variables/ Output var
y.head()
```

```python
x=df.iloc[:,:-1] # dependent Variables/ Input Var
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
x_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
sc=StandardScaler() #It is used to resize the distribution of Values so that the mean of the observed value is 0 and Standard deviation is 1.
#x_train=sc.fit_transform(x_train)
#x_test=sc.fit(x_test)
```

```python
x_train
```

```python
pca=PCA() #Principal Component Analysis: It's a technique to reduce the data in 
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)
```

```python
from sklearn.decomposition import FactorAnalysis as ft
```

```python
fa=ft(n_components=1)
x_train=fa.fit_transform(x_train,y_train)
x_test=fa.transform(x_test)
```

```python
lda=Lda(n_components=1) # used for multi class classification that can be used to automatically perform the dimensionality reduction.
x_train=lda.fit_transform(x_train,y_train)
x_test=lda.transform(x_test)
```
