```python
from apyori import apriori
```

```python
churn=pd.read_csv('New Retail.csv',header=None)
```

```python
churn.head()
```

```python
churn.shape
```

```python
records=[]
for i in range(0,15):
    records.append([str(churn.values[i,j]) for j in range(0,6)])
```

```python
records[1]
```

```python
association_rules=aprori(records,min_support=0.0045,min_confidence = 0.2,min_lift=3,min_length=2)
association_results=list(association_rules)
```
