# **Create and Customize Bar Plot**
With Pyplot, you can use the bar() function to draw bar graphs.

The bar() function takes arguments that describes the layout of the bars.

The categories and their values represented by the first and second argument as arrays.

### Horizontal Bars
If you want the bars to be displayed horizontally instead of vertically, use the barh() function:

```python
# Create a Bar plot:- 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# create dataset:-
data = {"Apples":20,"Mangoes":15,"Lemons":30,"Oranges":18}
names = list(data.keys())
values = list(data.values())

# define figure size:
fig = plt.figure(figsize = (15,5))

# Create Bar plot 
plt.bar(names,values, color = "blue")
plt.show()
```

```python
# Create a Bar plot:- 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# create dataset:-
data = {"Apples":20,"Mangoes":15,"Lemons":30,"Oranges":18}
names = list(data.keys())
values = list(data.values())

# define figure size:
fig = plt.figure(figsize = (10,5))

# Create and customize  Bar plot 
plt.bar(names,values, color = "navy") #for change bar color
plt.title("Bar Graph")
plt.xlabel("Name of the Fruits")
plt.ylabel("Quantity")
plt.grid(False)

plt.show()
```

```python
# create Horizontally bar plot:-
plt.barh(names,values, color = "blue") #for change bar color
plt.show()
```

# **Scatter plot:-**
With Pyplot, you can use the scatter() function to draw a scatter plot.

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
#Create a Scatter plot:-
# import the required library:
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

#create Datasets:
a = [10,20,30,14,35,5,43,13]
b = [5,2,3,4,8,6,9,7]
x = [1,2,3,4,5,6,7,8]

#Create the scatter plot:
plt.scatter(a,b)
plt.scatter(a,x)
plt.show()


```

```python
# customize a Scatter plot:-
# import the required library:
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

#create Datasets:
a = [10,20,30,14,35,5,43,13]
b = [5,2,3,4,8,6,9,13]
x = [1,2,3,4,5,6,7,78]

#Create the scatter plot:
plt.scatter(a,b, c="y", s = 200, edgecolors="b",marker='o',alpha=0.5)
plt.scatter(a,x, c="r", s = 200, edgecolors="g",marker='*',alpha=0.5)
plt.legend(['b','x'], loc = "best")
plt.title("Scatter plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.savefig("Scatter.png") # for  save the figure
plt.grid(True)
plt.show()
```

# **Histogram plot**
A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

In Matplotlib, we use the hist() function to create histograms.

The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

```python
#create a Histogram plot:
import matplotlib.pyplot as plt
%matplotlib inline

#define a dataset
num = [1,2,3,10,15,20,22,26,30,35,39,40,42,43,45,50,58,60,63,65,70,71,72,78,8,85,89,90,91,92,93]

#create histogram plot by using the above dataset
plt.hist(num, bins =20) # bins is for range the value of histogram.
plt.show()

```

```python
# customize the Histogram plot:
plt.hist(num, bins = [0,10,20,35,40,55,60,70,80,90,100], edgecolor = "blue", color = "green")
plt.title("Histogram Plot")
plt.xlabel("Range of Values")
plt.ylabel("Frequency of the Values.")
plt.grid(False)
plt.savefig("Histogram.png")
plt.show()
```

# **Box Plot:-**
Boxplots are a measure of how well distributed the data in a data set is. It divides the data set into three quartiles. This graph represents the minimum, maximum, median, first quartile and third quartile in the data set. It is also useful in comparing the distribution of data across data sets by drawing boxplots for each of them.
Boxplot can be drawn calling Series.box.plot() and DataFrame.box.plot(), or DataFrame.boxplot() to visualize the distribution of values within each column.

```python
# Create a box plot:-
import matplotlib.pyplot as plt
%matplotlib inline 

#define Dataset;
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total,order,discount])

# Ploting the data into boxplot:
plt.boxplot(data, showmeans = True)
plt.title("Box Plot")
plt.grid(True)
plt.show
```

# Violin Plot:- 
These are sililar to **Box Plots** except that they also show the probablilty, density of the data at different values.

It includes a marker for the median of the data and a box indacating the interquartile range, as in the standard box plots.

It is more informative than a plain box, it shows the full distribution of the box.


```python
import matplotlib.pyplot as plt
%matplotlib inline 

#define Dataset;
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total,order,discount])

#Create violin plot:
plt.violinplot(data,showmeans = True, showmedians = True)
plt.title("Violin Plot")
plt.grid(True)
plt.show()
```

# **Pie plot:-**
1. It can display only one series of the data.
2. It shows the size ofitems called wedge in one data series, proportional to the sum of the items.
3. The data points in a pie chart are shown as percentage of the whole chart.


```python
# Create a pie plot:-
import matplotlib.pyplot as plt
%matplotlib inline 

# define dataset.
animal = ["Dog","Cat","Lion","Tiger"]
size = [50,45,60,80]

fig = plt.figure(figsize = (10,5))
#Create a piechart:-
#plt.pie(size,labels = animal)

# Customize the piechart:
plt.pie(size,labels = animal , shadow=True, startangle=120)
plt.title("Pie Chart")
plt.show()
```

```python
#To add a list of explanation for each wedge, use the legend() function:
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, shadow = True, startangle = 45)
plt.legend(loc = "best")
plt.show() 
```

```python
#Explode the data part 
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1,0,0,0.2]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```
