```python
import pandas as pd
```

```python
import numpy as np
```

```python
churn=pd.read_csv('diabetes-1.csv')
```

```python
churn.head()
```

```python
x=churn.iloc[:,0:4].values
```

```python
y=churn.iloc[:,4].values
```

```python
x.shape
```

```python
y.shape
```

```python
 from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
algo_obj=RandomForestRegressor()
```

```python
algo_obj.fit(x_train,y_train)
```

```python
y_pre=algo_obj.predict(x_train)
```

```python
y_train
```

```python
y_pre
```

```python



from sklearn.metrics import mean_squared_error
import numpy as np
np.sqrt(mean_squared_error(y_train,y_pre))
```
